```python
import pandas as pd
import numpy as np
from os import remove
from os import path
from numpy import pi, cos, sin, sqrt
from numpy import arccos as acos
import math
import matplotlib.pyplot as plt
from datetime import datetime
import streamlit as st
import altair as alt
from matplotlib import cm
import random 
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
import requests
import pydeck as pdk
import json
```

```python

latitud_f=47.7197
longitud_f=12.8928

url = "https://maps.googleapis.com/maps/api/elevation/json?locations="+str(latitud_f)+","+str(longitud_f)+"&key=AIzaSyDhpSAZtcnr-OvNY1f58WpMMXFG_V1_lSs"

payload={}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
results=json.loads(response.text)["results"]
if len(results)==0:
    print("error")
else:    
    elevation_f=int(results[0]['elevation'])
```

```python
lat_f=np.radians(latitud_f)
lon_f=np.radians(longitud_f)

del_lat_grad=0.0035
del_lon_grad=0.0035
dis=150
dis_lat=0
dis_lon=0
Rt=6371000
i=0
while not(((dis_lat>dis*0.999)and(dis_lat<dis*1.001)) and ((dis_lon>dis*0.999)and(dis_lon<dis*1.001))):
    
    if dis_lat>dis:
        del_lat_grad-=0.000001
    else:
        del_lat_grad+=0.000001
    if dis_lon>dis:
        del_lon_grad-=0.000001
    else:
        del_lon_grad+=0.000001
    del_lat=np.radians(del_lat_grad)
    del_lon=np.radians(del_lon_grad)
    
    dis_lat=Rt*acos(sin(lat_f)*sin(lat_f+del_lat)+cos(lat_f)*cos(lat_f+del_lat)*cos(0))
    dis_lon=Rt*acos(sin(lat_f)*sin(lat_f)+cos(lat_f)*cos(lat_f)*cos(del_lon))
print("dis_lat",dis_lat)
print("dis_lon",dis_lon)
    

del_lon_grad
```

```python
num_r=20
fac=10
t=np.power(np.linspace(0.5,fac,num_r),0.5 )
num_ang=20
ang=np.linspace(0,2*pi,num_ang)
r=np.linspace(dis,num_r*dis,num_r)*t
print(t,r)
```

```python
elev_p=np.array([elevation_f])
dis_lat_p=np.array([0])
dis_lon_p=np.array([0])
lat_p=np.array([latitud_f])
lon_p=np.array([longitud_f])
for ii, r_p in enumerate(r):
 
    for kk in ang:
        lat_ii=latitud_f+(t[ii])*(ii+1)*cos(kk)*del_lat_grad
        lon_kk=longitud_f+(t[ii])*(ii+1)*sin(kk)*del_lon_grad
        dis_lat_ii=r_p*cos(kk)
        dis_lon_kk=r_p*sin(kk)
        
        url = "https://maps.googleapis.com/maps/api/elevation/json?locations="+str(lat_ii)+","+str(lon_kk)+"&key=AIzaSyDhpSAZtcnr-OvNY1f58WpMMXFG_V1_lSs"
        response = requests.request("GET", url, headers=headers, data=payload)
        results=json.loads(response.text)["results"]
        if len(results)==0:
            print("error")
            eleviikk=0
        else:    
            eleviikk=int(results[0]['elevation'])
        dis_lat_p=np.append(dis_lat_p,dis_lat_ii)
        dis_lon_p=np.append(dis_lon_p,dis_lon_kk)
        lat_p=np.append(lat_p,lat_ii)
        lon_p=np.append(lon_p,lon_kk)
        elev_p=np.append(elev_p,eleviikk)
```

```python
%matplotlib notebook
fig, ax = plt.subplots(figsize=(10,6))
plt.title('Jueputa')
ax.plot(dis_lon_p, dis_lat_p)
ax.grid(True)

```
